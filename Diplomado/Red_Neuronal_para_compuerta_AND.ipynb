## RETO DE APRENDIZAJE 3 - Creación de red neuronal

**Objetivo: Desarrollar una red neuronal la cual realice la función AND.**

Para este código se usará Keras, que es la librería con la que podemos describir/especificar las capas (layers) y simplifica la implementación de redes neuronales.

Pasos de la implementación:

1. Definir la función de activación, en este caso la función sigmoide y relu.
2. Inicializar los parámetros del modelo
3. Ajustar el modelo
5. Evaluar el performace del modelo

## Librerías

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
and_elements = {"x1":[0,0,1,1],"x2":[0,1,0,1], "y":[0,0,0,1] }

and_df = pd.DataFrame(data=and_elements)
and_df
```

## Implementación

```python
# Paso 1 y 2: Definición de función de activación e inicialización de parámetros

x_and= and_df[["x1","x2"]]
y_and = and_df["y"]

# Va a poner las capas en secuencia
model = Sequential()

# En la primera capa hay dos neuronas de input y el hidden layer tiene 16 neuronas
# Add() nos regresa un solo tensor
model.add(Dense(16, input_dim=2,
                    activation='relu'))

# Para el output
model.add(Dense(1, activation='sigmoid'))
```

```python
# Paso 3: Ajustar el modelo

model.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])

# Mientras más completo sea el modelo se necesitan más interaciones

model.fit(x_and,y_and,epochs=250)
```

Para configurar el proceso de aprendizaje con el método de compile():
Esto se hace definiendo el optimizador, la función de pérdida y las métricas a seguir durante el entrenamiento.

**Loss:** Una función de pérdida, también conocida como función de costo, es un método utilizado para estimar las
discrepancias entre los valores reales y predichos en un modelo de aprendizaje automático. Es un componente clave en el proceso de entrenamiento de redes neuronales, ya que proporciona una medida del rendimiento del modelo.

**Optimizer:** Los optimizadores son algoritmos o métodos que se utilizan para cambiar o ajustar los atributos de una red neuronal, como los pesos de las capas, la tasa de aprendizaje, etc., para reducir la pérdida y, a su vez, mejorar el modelo.

**Metrics:** Mientras que la función de pérdida se utiliza para el entrenamiento, la métrica se utiliza para la evaluación. Proporciona una medida más interpretable del rendimiento de su modelo, ya que suele estar en una unidad que es más fácil de entender. Ej. una precisión de 0.85 significa que el modelo es correcto el 85% de las veces.

[Fuente](https://saturncloud.io/blog/understanding-the-difference-between-loss-function-and-metric-in-keras/)

**Epoch:** generalmente definido como una iteración sobre todo el conjunto de datos", se utiliza para separar el entrenamiento en distintas fases, lo cual es útil para el registro y la evaluación periódica.

```python
# Paso 4: Evaluación del modelo

scores = model.evaluate(x_and, y_and)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
